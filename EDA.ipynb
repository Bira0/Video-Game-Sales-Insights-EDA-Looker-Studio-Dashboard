# Data Wrangling

## Feature Engineering

Dropping column 'Rank' since it has no significane to our analysis.

```python
games.shape
```

```python
games = games.drop('Rank', axis=1)
```

Dropping rows with missing values in both 'Year' & 'Publisher'

```python
games = games.dropna(subset=['Year', 'Publisher'])
```

```python
games['Year'].max()
```



*   Creating feature 'Game_Age' #Feature extraction
*   Since this dataset was generated and last updated around 2020, the age will adhere to that year


```python
games['Game_Age'] = 2020 - games['Year']
```

* Creating feature sale percentage from each sales region contributing to global sales

```python
# percentage contribution of each region to Global Sales
games['NA Sales Percentage'] = (games['NA_Sales'] / games['Global_Sales']) * 100
games['EU Sales Percentage'] = (games['EU_Sales'] / games['Global_Sales']) * 100
games['JP Sales Percentage'] = (games['JP_Sales'] / games['Global_Sales']) * 100
games['Other Sales Percentage'] = (games['Other_Sales'] / games['Global_Sales']) * 100
```

* Creating a 'Sales Category' feautre
* Low, Medium, High

```python
def categorize_sales(sales):
  if sales < 1:
    return 'Low'
  elif sales <= 10:
    return 'Medium'
  else:
    return 'High'

games['Sales Category'] = games['Global_Sales'].apply(categorize_sales)
```

```python
games.shape
```

* Creating feature 'Publisher Global Sales'

```python
publisher_sales = games.groupby('Publisher')['Global_Sales'].sum()
games['Publisher Global Sales'] = games['Publisher'].map(publisher_sales)
games['Publisher Global Sales']
```

Most profitable Genre
* Creating feature 'Genre Global Sales' by calculating total sales for each genre

```python
genre_sales = games.groupby('Genre')['Global_Sales'].sum()
games['Genre Global Sales'] = games['Genre'].map(genre_sales)
games['Genre Global Sales']
```

# Exploratory Data Analysis 'EDA'

Global and Regional Sales Trends
* North America contributes the largest share of global sales.
* Europe follows with 27.33% of sales, while Japan has a significant share despite being a smaller market.

```python
import seaborn as sns
```

```python
regional_sales = games[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
regional_sales.plot(kind='pie', autopct='%1.2f%%', figsize=(10, 8), colors=['#876FD4', '#4394E5', '#87BB62', '#F5921B'])
plt.title('Regional Sales Distribution')
plt.ylabel('')
plt.show()
```

```python
regional_trends = games.groupby("Year")[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
regional_trends.plot(kind='bar', stacked=True, figsize=(12, 6), cmap='viridis', edgecolor='black')
plt.title('Regional Sales Trends Over Time')
plt.xlabel('Year')
plt.ylabel('Sales (Millions)')
plt.xticks(rotation=45)
plt.legend(title='Region')
plt.show()
```

Genre Popularity
* Sports, Action, and Shooter games lead in global sales
* Role-playing games are particularly popular in Japan
* Action dominating all regions except for Japan.

```python
genre_sales = games.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
genre_sales.plot(kind='bar', color='#5E40BE', edgecolor='black')
plt.title('Total Global Sales by Genre')
plt.xlabel('Genre')
plt.ylabel('Total Global Sales (Millions)')
plt.xticks(rotation=45)
plt.show()
```

```python
#Total sales of each region by genre
region_total_sales = games.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

plt.figure(figsize=(12, 6))
sns.heatmap(region_total_sales, annot=True, fmt=".1f", cmap='coolwarm', linewidths=0.5)
plt.title('Regional Sales by Genre (in Millions)')
plt.xlabel('Region')
plt.ylabel('Genre')
plt.show()
```

 Platform Success
* Platforms such as the Wii, PS2, PS3, and Xbox 360 dominate global sales, hosting a large number of successful games.
* Platform popularity has shifted significantly over the years, with newer platforms like PS4 gaining traction.


```python
# Horizontal bar plot for total global sales by platform
platform_sales = games.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=True)
plt.figure(figsize=(10, 8))
platform_sales.plot(kind='barh', color='#004D99', edgecolor='black')
plt.title('Total Global Sales by Platform', fontsize=16, fontweight='bold')
plt.xlabel('Global Sales (Millions)', fontsize=14)
plt.ylabel('Platform', fontsize=14)
plt.grid(axis='x', alpha=0.5, color='#A3A3A3')
plt.show()
```

```python
top5 = ['Wii', 'PS2', 'PS4', 'Xbox 360', 'DS']

filter = games[games['Platform'].isin(top5)]

plt.figure(figsize=(10, 6))
for platform in top5:
    data = filter[filter['Platform'] == platform].groupby('Year')['Global_Sales'].sum()
    plt.plot(data.index, data.values, marker='o', label=platform)

plt.title('Platform Sales Over Time')
plt.xlabel('Year')
plt.ylabel('Global Sales (Millions)')
plt.legend(title='Platform')
plt.grid(alpha=0.5)
plt.show()
```

Publisher Performance
* Nintendo leads global sales by a significant margin, followed by Electronic Arts.
* Most publishers experienced steady growth over time, while Nintendo stood out with significant bursts in sales during key periods, followed by notable declines.


```python
top_publishers = games.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
top_publishers.plot(kind='barh', color='#004D99', edgecolor='black')
plt.title('Top 10 Publishers by Global Sales')
plt.xlabel('Publisher')
plt.ylabel('Global Sales (Millions)')
plt.show()
```

```python
publisher_year_sales = games.groupby(['Year', 'Publisher'])['Global_Sales'].sum().reset_index()

top_publishers = publisher_year_sales.groupby('Publisher')['Global_Sales'].sum().nlargest(5).index
filtered_data = publisher_year_sales[publisher_year_sales['Publisher'].isin(top_publishers)]


plt.figure(figsize=(12, 6))
for publisher in top_publishers:
    data = filtered_data[filtered_data['Publisher'] == publisher]
    plt.plot(data['Year'], data['Global_Sales'], marker='o', label=publisher, linewidth=2)


plt.title('Sales Trends Over Time for Top 5 Publishers', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Global Sales (Millions)', fontsize=14)
plt.legend(title='Publisher', fontsize=12)
plt.grid(alpha=0.5)
plt.tight_layout()
plt.show()
```
